##### 6. Plot a correlation matrix, and comment on what you observe.

```python
# Storing only the numerical columns to a additional dataframe
numerical_df = customer_df.select_dtypes(include='number')

# Checking correlations with the total_claim_amount:
numerical_df.corr()['total_claim_amount'].sort_values(ascending=False)
```

------------------------------------------------------------------------------------------------------
The correlation matrix provides valuable insights into the relationships between pairs of numerical variables in your dataset. Here's how to interpret the resulting correlation matrix:

1. **Interpreting the values**:
   - The correlation coefficient ranges from -1 to 1, where:
       - 1 indicates a perfect positive linear relationship,
       - -1 indicates a perfect negative linear relationship,
       - 0 indicates no linear relationship.
   - The closer the value is to 1 or -1, the stronger the linear relationship.
   - A value close to 0 indicates no linear relationship.

2. **Specific interpretation**:
    - The variable `total_claim_amount` has a strong positive correlation (0.63) with `monthly_premium_auto`, meaning that as one variable increases, the other tends to increase as well.
    - There is a moderate positive correlation (0.23) between `total_claim_amount` and `customer_lifetime_value`.
    - The variable `income` has a moderate negative correlation (-0.36) with `total_claim_amount`, indicating that as income increases, the total claim amount tends to decrease. Conversely, as income decreases, the total claim amount tends to increase.

3. **Interpreting absolute values**:
   - Absolute values closer to 1 indicate stronger relationships. For example, a correlation coefficient of 0.63 is stronger than 0.23.

4. **Cautions**:
   - Correlation does not imply causation. While a correlation suggests a relationship between two variables, it does not prove that changes in one variable cause changes in the other.

By incorporating this interpretation, you can better understand the relationships between the numerical variables in your dataset.

-------------------------------------------------------------------------------------------------------

------------------------------------------------------------------------------------------------------
A positive correlation of 0.23 between the `total_claim_amount` and `customer_lifetime_value` indicates that there is a moderate positive linear relationship between these two variables in your dataset.

Interpreting a positive correlation of 0.23:
- As the `customer_lifetime_value` increases, there tends to be an increase in the `total_claim_amount`, and vice versa.
- However, the strength of this relationship is considered moderate, as the correlation coefficient is 0.23 (not approaching 1).

This positive correlation could suggest several potential explanations or relationships between these two variables:
- Customers with higher lifetime value might make more, or more costly, insurance claims.
- Alternatively, it might indicate that customers who make more insurance claims tend to have a longer lifetime value with the company.
